```python
#导入三维工具包mplot3d
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
#创建3d绘图区域
ax = plt.axes(projection='3d')

#从三个维度构建
k = 500
z = np.linspace(0, 1, 100)
x = z * np.sin(k * z)
y = z * np.cos(k * z)

#调用 ax.plot3D创建三维线图
ax.plot3D(x, y, z, 'gray')
ax.set_title('3D line plot')
plt.show()
```

```python
#导入三维工具包mplot3d
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
#创建3d绘图区域
ax = plt.axes(projection='3d')

dataset = np.array([[1, 1],
                    [2, 2],
                    [3, 3]])
# Hypothesis:
# h = b + kx
def Hypothesis(k, b, x):
    return b + k*x

#Cost Function
def costFunction(k,b):
    sum = 0
    count = 0
    for p in dataset:
        count+=1
        sum += (Hypothesis(k, b, p[0])-p[1])**2
    # print(sum)
    return sum/2/count


#从三个维度构建
#求向量积(outer()方法又称外积)
k = np.outer(np.linspace(-20, 20, 100), np.ones(100))
#矩阵转置
b = k.copy().T
z = costFunction(k,b)

#调用plot_surface()函数
ax.plot_surface(b, k, z,cmap='viridis', edgecolor='none')
ax.set_title('Find the lowest point of J($\Theta_0$,$\Theta_1$)')
ax.set_xlabel('$\Theta0$(b)')
ax.set_ylabel('$\Theta1$(k)')
ax.set_zlabel('J($\Theta_0$,$\Theta_1$)')
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
#求向量积(outer()方法又称外积)
x = np.outer(np.linspace(-2, 2, 30), np.ones(30))
#矩阵转置
y = x.copy().T 
#数据z
z = np.cos(x ** 2 + y ** 2)
#绘制曲面图
fig = plt.figure()
ax = plt.axes(projection='3d')
#调用plot_surface()函数
ax.plot_surface(x, y, z,cmap='viridis', edgecolor='none')
ax.set_title('Surface plot')
plt.show()
```
